```python
import ivolatility as ivol
import pandas as pd
import os

ivolApiKey = os.getenv('API_KEY')
ivol.setLoginParams(apiKey=ivolApiKey)

# Let us be interested in call options on "AAPL" with expiration dates up to 30 days and delta (Greek) between 0.5 and 0.8 on the date "2025-03-01"
getOptsData = ivol.setMethod('/equities/eod/stock-opts-by-param')

OptsData = getOptsData(symbol='AAPL',startDate='2025-03-01', endDate='2025-03-27',dteFrom=0,dteTo=30,deltaFrom=0.5,deltaTo=0.8,cp='C')
OptsData.reset_index(drop=True, inplace=True)

OptsData


# Let's give a similar example, but for the "BMW" put options and this time will use the “moneyness” (out-of-the-money) parameters.
# In this case we need to specify the Region = "Europe" parameter
# OptsData = getOptsData(symbol='BMW',startDate='2025-03-01', endDate='2025-03-27',dteFrom=0,dteTo=30,moneynessFrom=-50,moneynessTo=-10,cp='C',region='EUROPE')
# OptsData.reset_index(drop=True, inplace=True)

# OptsData
```
